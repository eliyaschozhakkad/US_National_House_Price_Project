```python
X = df.drop(columns = ['DATE','S&PCase_Schiller_Index'])
X.head()
```

```python
y = df['S&PCase_Schiller_Index']
y.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
X_train,X_test, y_train, y_test = train_test_split(X,y,test_size =0.20,random_state = 42)
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
lr.coef_
```

```python
X.columns
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
mean_squared_error(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

```python
from sklearn.linear_model import Lasso,Ridge,ElasticNet
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
from sklearn.ensemble import AdaBoostRegressor, BaggingRegressor, ExtraTreesRegressor, GradientBoostingRegressor, \
    RandomForestRegressor, StackingRegressor, VotingRegressor
```

```python
base_regressors = [
    ('rf', RandomForestRegressor(n_estimators=50, random_state=42)),
    ('gb', GradientBoostingRegressor(n_estimators=50, random_state=42)),
    ('svr', SVR())
]

final_regressor = LinearRegression()
```

```python
models = {
    "Linear_model":LinearRegression(),
    "Lasso_model":Lasso(),
    "Ridge_model":Ridge(),
    "ElasticNet_model":ElasticNet(),
    "DecisionTreeRegressor":DecisionTreeRegressor(),
    "SVR":SVR(),
    "AdaBoostRegressor":AdaBoostRegressor(),
    "BaggingRegressor":BaggingRegressor(),
    "ExtraTreesRegressor":ExtraTreesRegressor(),
    "GradientBoostingRegressor":GradientBoostingRegressor(),
    "RandomForestRegressor":RandomForestRegressor(),
    "StackingRegressor":StackingRegressor(estimators=base_regressors, final_estimator=final_regressor),
    "VotingRegressor":VotingRegressor(estimators=base_regressors)
}
```

```python
accuracy = {}
for item,value in models.items():
  model = value
  model.fit(X_train,y_train)
  y_pred = model.predict(X_test)
  r2_score_train = round(model.score(X_train,y_train)*100,2)
  r2_score_test = round(model.score(X_test,y_test)*100,2)
  mse = round(mean_squared_error(y_test,y_pred),2)
  mae = round(mean_absolute_error(y_test,y_pred),2)
  rmse = round(np.sqrt(mse),2)
  accuracy[item] = [r2_score_train,r2_score_test,mse,mae,rmse]


print(accuracy)
```

```python
result = pd.DataFrame(data =pd.DataFrame(accuracy).T)
result.rename(columns={0:"R2_Training_Accuracy",1:"R2_Testing_Accuracy",2:"Mean_squared_error",3:"Mean_absolute_error",4:"Root_mean_squared_error"},inplace = True)
result
```

## Best Models

1. ExtraTreesRegressor have highest accuracy of 99.75% with least RMSE 1.81

2. RandomForestRegressor have  accuracy of 99.40% with  RMSE of 2.80

3. GradientBoostingRegressor	have accuracy of 99.35% with RMSE of 2.93

4. LinearRegression have accuracy of  99.15% with RMSE of 3.35

### Conclusion

1.**If we prefer more accuracy with more computational cost,we can go for ensemble models like ExtraTreesRegressor,RandomForestRegressor or GradientBoostingRegressor.**

2.**If we prefer computational cost with some compromise in accuracy,we can go for LinearRegression model which will be faster than other ensemble methods.**
