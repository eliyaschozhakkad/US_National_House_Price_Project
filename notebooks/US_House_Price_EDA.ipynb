# Key factors affecting US Home Prices Nationally(S&PCase_Schiller_Index).

##The data were collected publically from Fred Economic data which is publicaly available.

1. Median_Sale_Price_House - https://fred.stlouisfed.org/series/MSPUS.

 -   *'The Median Sale Price of a House is the middle value of all house sale prices in a given area, providing a representative measure of central tendency.'*

2. Personal_Income_Per_Capita - https://fred.stlouisfed.org/series/A792RC0Q052SBEA

 -   *'
Personal Income Per Capita is the average income earned by an individual in a specific geographical area or region, calculated by dividing the total personal income by the population.'*


3. Property_Tax - https://fred.stlouisfed.org/series/S210400

 -   *'
Property Tax is a local government levy on the assessed value of real estate properties, typically imposed annually as a source of revenue for funding public services and infrastructure.'*

4. Nasdaqcom - https://fred.stlouisfed.org/series/NASDAQCOM

 -   *'NASDAQ is an American stock exchange known for its electronic trading platform, hosting a diverse array of companies, particularly in the technology sector.'*

5. Construction_Employment_Cost - https://fred.stlouisfed.org/series/CIU2012300000000I

 -   *'
Construction Employment Cost refers to the total expenses incurred by an employer in the construction industry to compensate and provide benefits to its workforce, encompassing wages, salaries, and associated employment-related costs.'*

6. Population - https://fred.stlouisfed.org/series/POPTHM

 -   *'Population refers to the total number of individuals living in a specific geographical area, such as a country, city, or region, at a given point in time.*

7. New_House_Units_Construction - https://fred.stlouisfed.org/series/UNDCONTSA

 -   *'
New House Units Construction refers to the process of building and creating new residential structures or dwellings, typically measured by the number of housing units under construction within a specified area or timeframe.*

8. Consumer Price Index(CPI) - https://fred.stlouisfed.org/series/CORESTICKM159SFRBATL

 -   *'
The Consumer Price Index (CPI) is a measure that examines the weighted average of prices of a basket of consumer goods and services, such as transportation, food, and medical care, over time, reflecting changes in the cost of living and inflation.'*

9. Mortage_Rate - https://fred.stlouisfed.org/series/MORTGAGE30US

 -   *'
Mortgage Rate refers to the interest rate charged by a lender on a mortgage loan, representing the cost of borrowing for the homebuyer. It is expressed as a percentage and plays a crucial role in determining the overall cost of financing a home purchase.'*

10. Unemployment_Rate - https://fred.stlouisfed.org/series/UNRATE

 -   *'
The Unemployment Rate is a measure that expresses the percentage of the labor force that is currently unemployed and actively seeking employment, providing an indicator of the overall health of the job market within a specific region or country..'*

11. Household_Debt_Payment - https://fred.stlouisfed.org/series/TDSP

 -   *
'Household Debt Payment typically refers to the total amount of money that households spend on repaying debts, including mortgages, credit card payments, and other loans. It is a measure of the financial obligations that households must meet regularly to service their existing debt.'*

12. Recession_Rate_Prob -  https://fred.stlouisfed.org/series/RECPROUSM156N

 -   *'
Recession Rate refers to a significant decline in economic activity across the economy, typically characterized by a decrease in gross domestic product (GDP), employment, and other economic indicators, lasting for an extended period.'*

13. S&PCase_Schiller_Index  - https://fred.stlouisfed.org/series/CSUSHPISA

 -   *'The S&P/Case-Shiller Index is a widely followed benchmark that tracks changes in the value of residential real estate in various metropolitan areas in the United States.'*



### Importing Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
##Reading the csv file
data = pd.read_csv("data/house_price.csv")
```

```python
df = data.copy()
```

```python
df.shape
#No of datapoints -96
#No of features -14
```

```python
#Data information-datatypes
df.info()
```

```python
#Checking for null values in the data
df.isnull().sum()
```

```python
# Viewing the null value record
df[df.isnull().any(axis = 1)]
```

```python
## Only last row is empty and most values are empty.Since no informatin is present, we drop the last record permanently.
df.dropna(inplace = True)
```

```python
df.shape
```

### Datetype correction

```python
df.info()
```

```python
#Converting date from string format to date format
df["DATE"] = pd.to_datetime(df['DATE'])
```

```python
df.info()
```

```python
## Separating string columns
object_columns = [col for col in df.columns if df[col].dtype == 'O']
object_columns
```

```python
#Convert string to numerical values and round to 2 decimal places
for col in object_columns:
  df[col] = pd.to_numeric(df[col])
  df[col] = df[col].round(2)
```

```python
df.info()
```

```python
df.head()
```

```python
numerical_columns = [col for col in df.columns if df[col].dtype == 'float64']
```

```python
## Round all numerical values to 2 decimal places
for col in numerical_columns:
  df[col] = df[col].round(2)
```

```python
df.head()
```

### Feature Engineering

```python
#Extracting Quarter value from date
df['Date_Quarter'] = df['DATE'].dt.quarter
```

```python
#Extracting year value from date
df['Date_Year'] = df['DATE'].dt.year
```

```python
X = df.drop(columns = ['S&PCase_Schiller_Index'])
y = df['S&PCase_Schiller_Index']
```

### S&P Case-Shiller Index Versus Factors Influencing House Prices

```python

# Create a 3x5 grid of subplots
fig, axes = plt.subplots(7, 2, figsize=(20, 60))
fig.subplots_adjust(hspace=0.5)

for i, col in enumerate(numerical_columns):
    row, col_num = divmod(i, 2)


    sns.regplot(x=X[col], y=y, ax=axes[row, col_num])

    axes[row, col_num].set_title(col)
    axes[row, col_num].set_xlabel(col)
    axes[row, col_num].set_ylabel('S&PCase_Schiller_Index')

plt.show()
```

## Factors Affecting US National House Price(S&PCase_Schiller_Index):

### Positive correlation:

1. We can observe that as **Median Sale Price Houses** increases overtime ,**S&PCase_Schiller_Index** also increaese.This is due to increase in sales price.

2. **As people starts to earn more income, they decide to buy their own home or buy houses as an asset**.This is indicated by increase in S&PCase_Schiller_Index when there there is increase in Personal Income Per Capita.

3. **Absolute Property tax have also increased** due to increase in house prices which lead to increase in S&PCase_Schiller_Index.

4. **As Stockmarket performed well,People were able to get good returns and buy houses**.Stockmarket indicates economy of the country.When Nasdaqcom increased,S&PCase_Schiller_Index also increased.

5.**Due to inflation, contruction cost increases** mainly Construction Employement Cost.As a result S&PCase_Schiller_Index also increased.

6. **As population increases,people tend to buy new houses**.So the demand for houses increases and  S&PCase_Schiller_Index increases.

8. **As Consumer Price Index(CPI) increases over time,Peole have to pay higher prices** for houses.So S&PCase_Schiller_Index increases.



### Negative Correlation:

1. **As US government decreased Mortage rate continuosly over time ,People were able to buy houses**.When Mortage rate decreased,S&PCase_Schiller_Index increased.

2. **When unemployement rate increased,People were not able to buy houses**.So S&PCase_Schiller_Index decreased.

3**. When Household Debt payment decreased over time** , People were   able to buy houses and S&PCase_Schiller_Index  increased.

4. **During recession time,people were struggling due to economic stability and unemployment**,So houses were not sold in good number and S&PCase_Schiller_Index decreased

### Reasons for abnormality in S&PCase_Schiller_Index

- The Great Recession refers to the economic downturn from 2007 to 2009 after the bursting of the U.S. housing bubble and the global financial crisis.

- S&PCase_Schiller_Index decreased to near to 145 point till 2012 and then started to increase.

- It is also the worst time of US Real estate from 2006 to 2012 where there were less construction of new houses and people were suffering financially.Eventhough Stockmarket recovered faster,US Real Estate started to  recover by 2012.
